```python
#Diagonal elements are perfectly correlated
#Lighter shades denote negative correlation and darker shades show positive correlation.
#This figure shows the correlation between each of the variables and with that of the target variable, based on the correlation matrix calculated earlier
```

```python
#Extracting the variables which have a significant correlation with the target variable
#Following fucntion will consider the correlation values in the Price column and return those values which are either >0.5 or <-0.5
```

```python
corrmat.index
```

```python
def CorrFeat(corrdata,threshold):
    feature=[]
    value=[]

    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index])>threshold:
            feature.append(index)
            value.append(corrdata[index])
    
    corr_df=pd.DataFrame(data=value, index=feature, columns=['corr_value'])
    
    return corr_df
```

```python
threshold=0.5
corr_df = CorrFeat(corrmat['Price'], threshold)
```

```python
corr_df
```

```python
# Only RM shows a strong positive correlation with the Price. Higher the price, greater will be the average number of rooms per dwelling
# PTRATIO shows a strong negative correlation with Price. Higher the price, lower the pupil teacher ratio (i.e. more no. of teachers per student in town)
# LSTAT shows a strong negative correlation with Price. Higher the price, lower the % of lower status of the population (i.e. more upper class residents)
```
